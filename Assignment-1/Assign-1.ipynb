```python
import numpy as np
```

```python
# Q1

# (a) Reverse the array
arr = np.array([1, 2, 3, 6, 4, 5])
print("Reversed:", arr[::-1])

# (b) Flatten the array
array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])
print("Flatten using ravel:", array1.ravel())
print("Flatten using flatten:", array1.flatten())

# (c) Compare arrays
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
print("Arrays equal:", np.array_equal(arr1, arr2))

# (d) Most frequent value and indices
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1,1,1,2,3,4,2,4,3,3])

def most_frequent(arr):
    values, counts = np.unique(arr, return_counts=True)
    max_val = values[np.argmax(counts)]
    indices = np.where(arr == max_val)[0]
    return max_val, indices

print("X most frequent:", most_frequent(x))
print("Y most frequent:", most_frequent(y))

# (e) Sum operations on matrix
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Sum of all:", np.sum(gfg))
print("Row-wise sum:", np.sum(gfg, axis=1))
print("Column-wise sum:", np.sum(gfg, axis=0))

# (f) Matrix operations
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
print("Diagonal sum:", np.trace(n_array))
eigenvalues, eigenvectors = np.linalg.eig(n_array)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
print("Inverse:\n", np.linalg.inv(n_array))
print("Determinant:", np.linalg.det(n_array))

```

```python


# (g) Matrix multiplication and covariance

# --- First pair ---
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])

print("Matrix product 1:\n", np.dot(p1, q1))  # or p1 @ q1
print("Covariance 1:\n", np.cov(p1.T, q1.T))  # same shape â†’ works


# --- Second pair ---
p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])

print("Matrix product 2:\n", np.dot(p2, q2))

# Covariance between flattened versions (same length)
print("Covariance 2 (flattened):\n", np.cov(p2.flatten(), q2.flatten()))


# (h) Inner, outer, cartesian product
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

print("Inner product:\n", np.inner(x, y))
print("Outer product:\n", np.outer(x.flatten(), y.flatten()))

# Cartesian product
from itertools import product
cartesian = np.array(list(product(x.flatten(), y.flatten())))
print("Cartesian product:\n", cartesian)
```

```python
# Q2

# (a)
array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Absolute value
print("Absolute:\n", np.abs(array))

# ii. Percentiles
flat = array.flatten()
print("Flattened 25th, 50th, 75th:", np.percentile(flat, [25, 50, 75]))
print("Column percentiles:\n", np.percentile(array, [25, 50, 75], axis=0))
print("Row percentiles:\n", np.percentile(array, [25, 50, 75], axis=1))

# iii. Mean, Median, Std Dev
print("Flattened mean/median/std:", flat.mean(), np.median(flat), flat.std())
print("Column-wise mean/median/std:", array.mean(axis=0), np.median(array, axis=0), array.std(axis=0))
print("Row-wise mean/median/std:", array.mean(axis=1), np.median(array, axis=1), array.std(axis=1))

# (b)
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print("Floor:", np.floor(a))
print("Ceil:", np.ceil(a))
print("Trunc:", np.trunc(a))
print("Rounded:", np.round(a))
```

```python
#Q3

# (a)
array = np.array([10, 52, 62, 16, 16, 54, 453])
print("Sorted array:", np.sort(array))
print("Indices of sorted:", np.argsort(array))
print("4 smallest:", np.sort(array)[:4])
print("5 largest:", np.sort(array)[-5:][::-1])

# (b)
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
is_integer = array == array.astype(int)
print("Integer elements:", array[is_integer])
print("Float elements:", array[~is_integer])
```
